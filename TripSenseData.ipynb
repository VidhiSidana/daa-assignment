```python
import numpy as np
np.random.seed(42)
customer_agg['target_booking'] = np.random.choice([0,1], size=len(customer_agg), p=[0.3,0.7])
customer_agg['target_response'] = np.random.choice([0,1], size=len(customer_agg), p=[0.4,0.6])

```

```python
augmented_data=customer_agg.copy()
```

```python
numeric_cols = ['Age','NatureScore','CultureScore','AdventureScore','SpiritualScore',
                'BookingLeadDays','DurationDays','FamilySize','InternetSpeedMbps',
                'destination_diversity','category_diversity',
                'adventure_pref_score','culture_pref_score','nature_pref_score',
                'spiritual_pref_score','engagement_score']


np.random.seed(42)
noise_factor=0.05
synthetic_rows=augmented_data.copy()
for col in numeric_cols:
    noise = np.random.normal(0, noise_factor, size=len(synthetic_rows))
    synthetic_rows[col] = synthetic_rows[col] * (1 + noise)
synthetic_rows[numeric_cols] = synthetic_rows[numeric_cols].clip(0, 1)
```

```python
augmented_data = pd.concat([augmented_data, synthetic_rows], ignore_index=True)
print("Original rows:", len(customer_agg))
print("Augmented rows:", len(augmented_data))
```

```python
augmented_data = augmented_data.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
cluster_features = [
    'Age',
    'NatureScore', 'CultureScore', 'AdventureScore', 'SpiritualScore',
    'destination_diversity', 'category_diversity',
    'BookingLeadDays', 'DurationDays',
    'FamilySize',
    'InternetSpeedMbps',
    'engagement_score',
    'adventure_pref_score', 'culture_pref_score',
    'nature_pref_score', 'spiritual_pref_score'
]

X = augmented_data[cluster_features]
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
K = range(2, 11)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X)
    inertia.append(kmeans.inertia_)

plt.plot(K, inertia, marker='o')
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.title("Elbow Method")
plt.show()
```

```python
from sklearn.metrics import silhouette_score

sil_scores = {}

for k in range(2, 11):
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)
    sil_scores[k] = silhouette_score(X, labels)

sil_scores
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
augmented_data['cluster_kmeans'] = kmeans.fit_predict(X)
```

```python
from sklearn.cluster import AgglomerativeClustering

agg = AgglomerativeClustering(n_clusters=4)
augmented_data['cluster_hierarchical'] = agg.fit_predict(X)
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.7, min_samples=5)
augmented_data['cluster_dbscan'] = dbscan.fit_predict(X)
```

```python
from sklearn.metrics import silhouette_score

print("KMeans:", silhouette_score(X, augmented_data['cluster_kmeans']))
print("Hierarchical:", silhouette_score(X, augmented_data['cluster_hierarchical']))
```

```python
cluster_profile = augmented_data.groupby('cluster_kmeans')[cluster_features].mean()
cluster_profile
```

```python
augmented_data['engagement_score'] = (
    augmented_data['destination_diversity'] +
    augmented_data['category_diversity'] +
    (1 / (1 + augmented_data['BookingLeadDays']))
)
```

```python
augmented_data['travel_diversity_score'] = (
    augmented_data['destination_diversity'] * 0.6 +
    augmented_data['category_diversity'] * 0.4
)
```

```python
augmented_data['luxury_preference_score'] = (
    augmented_data['CultureScore'] * 0.4 +
    augmented_data['SpiritualScore'] * 0.3 +
    augmented_data['DurationDays'] * 0.3
)
```

```python
augmented_data['budget_score'] = (
    augmented_data['BookingLeadDays'] * 0.5 +
    (1 / augmented_data['DurationDays']) * 0.5
)
```

```python
augmented_data['weekend_score'] = (
    augmented_data['DurationDays'] <= 3
).astype(int)
```

```python
y = augmented_data['target_booking']
X_pred = augmented_data[cluster_features + [
    'engagement_score',
    'travel_diversity_score',
    'luxury_preference_score',
    'budget_score'
]]
```

```python
def recommend_destination(row):
    if row['AdventureScore'] > 0.7 and row['NatureScore'] > 0.6:
        return "Himalayan Treks, Forest Retreats"
    elif row['CultureScore'] > 0.7:
        return "Heritage Cities, Museums"
    elif row['SpiritualScore'] > 0.7:
        return "Pilgrimage Destinations"
    elif row['budget_score'] > 0.6:
        return "Budget Weekend Getaways"
    else:
        return "Popular Tourist Destinations"

augmented_data['recommended_trip'] = augmented_data.apply(recommend_destination, axis=1)
```
